```python
df_diabetes = pd.read_csv("Dataset of Diabetes .csv")
df_diabetes.head()
```

```python
import pandas as pd
import numpy as np

cols_with_zero = ['Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
df_diabetes[cols_with_zero] = df_diabetes[cols_with_zero].replace(0, np.nan)
for cols in cols_with_zero:
    df_diabetes[cols].fillna(df_diabetes[cols].mean(), inplace=True)

print("Missing values after cleaning:")
print(df_diabetes.isnull().sum())
```

```python

print("Shape before removing outliers:", df_diabetes.shape)
numeric_df = df_diabetes.select_dtypes(include=[np.number])

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)
df_diabetes = df_diabetes[~outliers]

print("Shape after removing outliers:", df_diabetes.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

numeric_cols = df_diabetes.select_dtypes(include=[np.number]).columns

scaler = MinMaxScaler()
df_diabetes_scaled = df_diabetes.copy()
df_diabetes_scaled[numeric_cols] = scaler.fit_transform(df_diabetes[numeric_cols])

print(df_diabetes_scaled.head())
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = df_diabetes.select_dtypes(include=[np.number]).columns

std_scaler = StandardScaler()
df_diabetes_standardized = df_diabetes.copy()
df_diabetes_standardized[numeric_cols] = std_scaler.fit_transform(df_diabetes[numeric_cols])

print(df_diabetes_standardized.head())
```
